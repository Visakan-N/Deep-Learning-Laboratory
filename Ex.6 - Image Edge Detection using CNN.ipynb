# Image Edge Detection:
Detecting the edges of an image by applying a kernal to it.

## Canny:

### Import the required packages:

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Load the image:

```python
img = cv2.imread('butterfly.jpg')
```

### Coverting to grayscale:
This improves the edge detection process.

```python
# converting the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

### Applying a blur function:
Using the Gaussian blur function to remove as much of the noise as possible in the image.

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0) # the second parm is the size of the kernal
```

### Applying the Canny function:
The range of the threshold can be altered. The range is directly proprtional to the number of edges

```python
canny = cv2.Canny(img, threshold1=180, threshold2=200) # wider the threshold range, more the edges
```

```python
fig, axs = plt.subplots(2, 2)
titles = ['Original', 'Grayscale', 'Gaussian Blur', 'Canny']
axs[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title(titles[0])
axs[0, 0].axis('off')
axs[0, 1].imshow(cv2.cvtColor(gray, None))
axs[0, 1].set_title(titles[1])
axs[0, 1].axis('off')
axs[1, 0].imshow(cv2.cvtColor(blur, None))
axs[1, 0].set_title(titles[2])
axs[1, 0].axis('off')
axs[1, 1].imshow(cv2.cvtColor(canny, None))
axs[1, 1].set_title(titles[3])
axs[1, 1].axis('off')
```

##  Sobel:

Repeat the same steps uptill the blur function from the canny algorithm.

### Applying the Sobel function:
ddepth - specifies the precision of the output image.

```python
sobelx = cv2.Sobel(src=blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3) # along the X axis
sobely = cv2.Sobel(src=blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3) # along the Y axis
sobelxy = cv2.Sobel(src=blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3) # Combined X and Y axes
```

```python
fig, axs = plt.subplots(2, 2)
titles = ['Original', 'Sobel along X-axis', 'Sobel along Y-axis', 'Sobel along both X and Y axis']
axs[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title(titles[0])
axs[0, 0].axis('off')
axs[0, 1].imshow(sobelxy, cmap='gray')
axs[0, 1].set_title(titles[3])
axs[0, 1].axis('off')
axs[1, 0].imshow(sobelx, cmap='gray')
axs[1, 0].set_title(titles[1])
axs[1, 0].axis('off')
axs[1, 1].imshow(sobely, cmap='gray')
axs[1, 1].set_title(titles[2])
axs[1, 1].axis('off')
```

## Prewitt:

### Defining the kernels:

```python
kernelx = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
kernely = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
```

###  Applying the kernels:
Convolutional 2D layers are used to find the edge of the image by passing the kernel to it.

```python
img_prewittx = cv2.filter2D(blur, -1, kernelx)
img_prewitty = cv2.filter2D(blur, -1, kernely)
img_prewitt = img_prewittx + img_prewitty
```

```python
fig, axs = plt.subplots(2, 2)
titles = ['Original', 'Kernel X', 'Kernel Y', 'Kernel X and Y']
axs[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title(titles[0])
axs[0, 0].axis('off')
axs[0, 1].imshow(img_prewitt, cmap='gray')
axs[0, 1].set_title(titles[3])
axs[0, 1].axis('off')
axs[1, 0].imshow(img_prewittx, cmap='gray')
axs[1, 0].set_title(titles[1])
axs[1, 0].axis('off')
axs[1, 1].imshow(img_prewitty, cmap='gray')
axs[1, 1].set_title(titles[2])
axs[1, 1].axis('off')
```
